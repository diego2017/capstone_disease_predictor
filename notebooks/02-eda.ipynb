### Imports

```python
# data manipulation
import numpy as np
import pandas as pd

# plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read csv file and provide overview of data

```python
# Load csv file
df_cleaned= pd.read_csv("../Data/Cleaned data/Diabetes_Health_Indicators_Cleaned_.csv")
```

## Exploring data

```python
df_cleaned.describe().T
```

```python
df_cleaned.info()
```

### Target variable
The target variable is the Has_diabetes column. As shown in the results below, there are 3 possible answer values found: 
* 0 -> When the person interviewed does not have diabetes or prediabetes
* 1 -> When the person interviewed has prediabetes
* 2 -> When the person interviewed has diabetes

Based on initial observations, 84% of the participants do not have diabetes or prediabetes, 14% have diabetes, and 1% have prediabetes.

There are no missing values in this column.

```python
Has_diabetes_counts = df_cleaned['Has_diabetes'].value_counts()
percentages = (Has_diabetes_counts / len(df_cleaned['Has_diabetes'])) * 100
print(percentages)
```

### Univariate analysis of all columns

```python
#count continous variables
cont_var=list(df_cleaned.select_dtypes(exclude='object'))

# Set up grid of subplots
###
plt.subplots(11,2,figsize=(15,8))

#Initialize count
plot_num=1

#Loop through column names in the list 
for col in cont_var:
    # Save column values, mean, and median to variables
    values_temp=df_cleaned[col]
    mean_temp=values_temp.mean()
    median_temp=values_temp.median()

    # Plot a histogram of song tempo, with lines marking mean and median
    plt.subplot(11,2,plot_num)
    sns.histplot(values_temp)
    plt.axvline(mean_temp, color='r', label='Mean')
    plt.axvline(median_temp, color='g',label='Median')
    plt.legend()

    plot_num+=1

plt.tight_layout()

plt.show() 
```

### Bivariate Analysis

```python
# Create variables for possible predictive columns and target variable
x_options=list(df_cleaned.select_dtypes(include='float'))
y=df_cleaned['Has_diabetes']
```

```python
# Create visualisations to show relationship between target variable and possible features
plt.subplots(10,2,figsize=(15,15))

plot_num=1

plt.suptitle("Relationship between someone's diabetes status and their other characteristics", fontsize=16)

for col in x_options:

    plt.subplot(10,2,plot_num)
    sns.scatterplot(x=df_cleaned[col],y=y, alpha=0.01)
    plt.title(f'Has Diabetes vs {col}')

    plot_num+=1

plt.subplots_adjust(top=0.93)

plt.show()
```
