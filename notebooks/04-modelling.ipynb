### Imports

```python
# data manipulation
import numpy as np
import pandas as pd

# plotting
import matplotlib.pyplot as plt
import seaborn as sns

# modelling
import statsmodels.api as sm
```

### Read csv file and provide overview of data

```python
# Load csv file
df_preprocessed= pd.read_csv("../Data/Prepocessed_data/Diabetes_Health_Indicators_Preprocessed.csv")
df_preprocessed.head()
```

## Modelling

```python
X = df_preprocessed.drop('Has_diabetes', axis=1).copy()
y = df_preprocessed['Has_diabetes'].copy()

print(f"Shape of feature matrix: {X.shape}")
print(f"Shape of target matrix: {y.shape}")
```

```python
# Start feature selection process
# write a reusable function to train a logistic regression model
def train_logistic_model(X, y):

    # 0. Add the constant
    X_const = sm.add_constant(X)

    # 1. Instantiate the model
    diabetes_logit = sm.Logit(y, X_const)  

    # 2. Fit the model to the data
    diabetes_logit_fitted = diabetes_logit.fit()

    # 3. Look at the results / summary table / assess model fit
    print(f'Summary of fitted logistic regression model: \n {diabetes_logit_fitted.summary()}')

    # 4. Calculate soft predictions
    y_proba = diabetes_logit_fitted.predict(X_const)

    # 5. Convert soft predictions to hard predictions 0/1
    y_pred = np.where(y_proba >= 0.5, 1, 0)

    # 6. Calculate #correct
    num_correct = (y_pred == y).sum()

    # 7. Calculate the percentage accuracy
    pct_accuracy = num_correct/X.shape[0]

    print(f'The baseline model accuracy is: {np.round(pct_accuracy*100.0, 3)}%')
    
    return diabetes_logit_fitted, pct_accuracy
```

```python
logit_fitted, pct_accuracy = train_logistic_model(X, y)
```

```python

sorted_coef = logit_fitted.params.sort_values()

#get the top and bottom coef
low_5_coef = sorted_coef.head(5)
high_5_coef = sorted_coef.tail(5)

# Join them
combined_coef = pd.concat([low_5_coef, high_5_coef])

# Display results
print("Top 5 coefficients with the lowest values and top 5 with the highest values are:")
print(combined_coef)
```

```python
#create a visualisation:
colour = ['red' if x < 0 else 'green' for x in combined_coef]
plt.figure(figsize=(10,5))
combined_coef.plot(kind='barh', color=colour)
plt.show()
```

Let's build a bar chart based on this data. We will convert coefficients into odds increases and decreases, color them in red and blue based on increase or decrease, and change the saturation based on p-values.

```python
coefficients_df = pd.DataFrame({"coef": logit_fitted.params,
                                "p_value": logit_fitted.pvalues})
coefficients_df.reset_index(inplace=True)


coefficients_df.rename({'index':'variable'}, axis='columns', inplace=True)
coefficients_df.head()

categories = []    # Holds labels for bars
sizes = []         # Hold bar heights
colors = []        # Holds bar colors
p_values = []      # Used for additional text within the plot

p_value_color_scale = 1.5

# iterate through
for index, row in coefficients_df.iterrows():
    variable_name = row['variable'].strip()
    
    if (variable_name != 'const'):
        categories.append(variable_name)
        coefficient_value = row['coef']
        
        p_value = round(row['p_value'],2)
        p_values.append(p_value)

        color = [1.0,1.0,1.0]
        
        color[1] = (min(p_value_color_scale*p_value, 1.0)/1.0)  # Make GREEN brighter
        
        if (coefficient_value >= 0):
            color[0] = color[1]                     # Make RED brighter
            column_size = np.exp(coefficient_value)
        else:
            color[2] = color[1]                     # Make BLUE brighter
            column_size = -1/np.exp(coefficient_value)

        column_size = round(column_size,2)
        sizes.append(column_size)
        colors.append(color)
```

```python
fix, ax = plt.subplots(figsize=(16, 12))
ax.barh(categories, sizes, color=colors)
for index, bar_size in enumerate(sizes):
    if (bar_size > 0.0):
        # Add a label text
        ax.text(bar_size + 0.5, index-0.15, f'{bar_size} odds increase', color='blue')
        
        # Add a p-value disclaimer
        if (p_values[index] >= 0.05):
            ax.text(-2.5, index-0.15, f'p-value={p_values[index]}', color='blue')
    else:
        # Label text
        ax.text(bar_size - 3.7, index-0.15, f'{abs(bar_size)} odds decrease', color='red')
        
        # p-value disclaimer
        if (p_values[index] >= 0.05):
            ax.text(0.3, index - 0.15, f'p-value={p_values[index]}', color='red')

# change the limit to make sure labels go inside the plot area
plt.xlim(-7.5, 16.5)
plt.title("Effect of different variables on risk of developing diabetes", size=20)
plt.xlabel("Odds change for a given feature", size=20)
plt.ylabel("Variable of interest", size=20)
plt.show()
```
