```python
# data manipulation
import numpy as np
import pandas as pd

# plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read csv file and provide overview of data

```python
# Load csv file
df_cleaned= pd.read_csv("../Data/Cleaned data/Diabetes_Health_Indicators_Cleaned_.csv")
```

### Target variable
The target variable is the Has_diabetes column. As shown in the results below, there are 3 possible answer values found: 
* 0 -> When the person interviewed does not have diabetes or prediabetes
* 1 -> When the person interviewed has prediabetes
* 2 -> When the person interviewed has diabetes

Based on initial observations, 84% of the participants do not have diabetes or prediabetes, 14% have diabetes, and 1% have prediabetes.

There are no missing values in this column.

```python
Has_diabetes_counts = df_cleaned['Has_diabetes'].value_counts()
percentages = (Has_diabetes_counts / len(df_cleaned['Has_diabetes'])) * 100
print(percentages)
```

### Univariate and bivariate analysis of all columns

```python
# Create lists of which columns are categorical and which ones continous
non_categorical_features = ['BMI', 'Mental_health_bad_days', 'Physical_health_bad_days']
categorical_features = [col for col in df_cleaned.columns if col not in non_categorical_features]
print(categorical_features)
```

```python
plt.subplots(1, len(non_categorical_features), figsize=(15, 5))

# Initialize count
plot_num = 1

# Loop through column names in the list of continuous features
for col in non_categorical_features:
    # Save column values, mean, and median to variables
    values_temp = df_cleaned[col]
    mean_temp = values_temp.mean()
    median_temp = values_temp.median()

    # Plot a histogram of the continuous feature, with lines marking mean and median
    plt.subplot(1, len(non_categorical_features), plot_num)
    sns.histplot(values_temp)
    plt.axvline(mean_temp, color='r', label='Mean')
    plt.axvline(median_temp, color='g', label='Median')
    plt.title(f'{col}')
    plt.legend()

    plot_num += 1

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(7, 3, figsize=(16, 30))  # Adjusted figure size

# Flatten the 2D array of subplots into a 1D array
axes = axes.ravel()

# Loop through categorical features
for i, column in enumerate(categorical_features):
    
    ax = axes[i]
    
    # Create a stacked bar chart
    df_cleaned.groupby([column, 'Has_diabetes']).size().unstack().plot(kind='bar', stacked=True, ax=ax)

    # Set labels and title
    ax.set_ylabel('Count')
    ax.set_xlabel(f'{column}')
    ax.set_title(f'Stacked Bar Chart for {column}')

# Remove any empty subplots
for i in range(len(categorical_features), 7 * 3):
    fig.delaxes(axes[i])

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Correlations between variables

```python
# Generate a correlation matrix
df_cleaned.corr()
```

```python
# Plot the correlation matrix as a heatmap to make it more readable

# mask for upper-triangle
mask = np.triu(df_cleaned.corr())

plt.figure(figsize=(15, 15))
sns.heatmap(df_cleaned.corr(), annot=True, center=0, cmap='coolwarm', vmin=-1, vmax=1, mask=mask)
plt.show()
```

Observations:
Based on the correlation matrix and heatmap above, we have detected multicollinearity that has to be fixed before creating the model during the next stage of the process.
Physical_health_bad_days and General_health_score have a high correlation, with 0.52, and will need to be deleted during pre-processing
